## Simple Thresholding
#### The simplest approach to segment an image is using thresholding. If pixel value is greater than a threshold value, it is assigned one value (may be white), else it is assigned another value (may be black).
Reference: http://www.cse.iitd.ernet.in/~pkalra/col783-2017/Thresholding.pdf


```python
import cv2
import numpy as np
from matplotlib import pyplot as plt


img = cv2.imread('standard_test_images/scrr.png',0) 
  
# calculate frequency of pixels in range 0-255 
histg = cv2.calcHist([img],[0],None,[256],[0,256])
plt.imshow(img,cmap="gray")
plt.show()
```

### If f(x,y)>T then f(x,y)=0   else f(x,y)=255

```python
res=np.where(img>48,0,255) #Taking threshold as 48
plt.imshow(res,cmap="gray")
plt.show()
```

```python
res=np.where(img>135,0,255) #Taking threshold as 135
plt.imshow(res,cmap="gray")
plt.show()
```

```python
#Histogram plots of the above image
plt.hist(img.ravel(),256,[0,256]) 
plt.show() 
```

```python
plt.plot(histg) 
plt.show() 
```

## Automatic thresholding
#### To make segmentation more robust, the threshold should be automatically selected by the system.
### - Choosing the theshold using the image histogram
#### Regions with uniform intensity give rise to strong peaks in the histogram

```python
img2 = cv2.imread('standard_test_images/pic2.png',0)
plt.imshow(img2,cmap = 'gray')
plt.show()
```

```python
#Lets plot the histogram
plt.hist(img2.ravel(),256,[0,256]) 
plt.show() 
```

```python
res2 = np.where(img2>=50,255,0) #segmenting out the circle object using thresholding
plt.imshow(res2,cmap='gray')
plt.show()
```

```python
res3 = np.where(img2>=170,255,img2) #segmenting out the squares using thresholding
plt.imshow(res3,cmap='gray')
plt.show()
```

```python
res4 = np.where(img2>=170,255,0) #segmenting out the squares using thresholding
plt.imshow(res4,cmap='gray')
plt.show()
```

## Adaptive Thresholding:
#### In this, the algorithm calculate the threshold for a small regions of the image. So we get different thresholds for different regions of the same image and it gives us better results for images with varying illumination.
### Adaptive Method - It decides how thresholding value is calculated.
Reference: https://opencv-python-tutroals.readthedocs.io/en/latest/py_tutorials/py_imgproc/py_thresholding/py_thresholding.html

```python
th2 = cv2.adaptiveThreshold(img2,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,11,2) 
#threshold value is the mean of neighbourhood area
plt.imshow(th2,cmap='gray')
plt.show()
```

```python
th3 = cv2.adaptiveThreshold(img2,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,11,2)
#threshold value is the weighted sum of neighbourhood values where weights are a gaussian window.
plt.imshow(th3,cmap='gray')
plt.show()
```
