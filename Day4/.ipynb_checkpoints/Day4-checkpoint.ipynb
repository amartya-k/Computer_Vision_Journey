### 1.1 Dilation:
#### Dilation expands the connected sets of 1s of a binary image.
#### It can be used for :
#### -growing features
#### -filling holes and gaps

```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
img = cv2.imread('im.png',0)
(thresh, bw) = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY) #Creating black and white image
plt.imshow(bw,cmap='gray')
plt.show()
```

```python
kernel = np.ones((8,8),np.uint8)  # a pixel element is '1' if atleast one pixel under the kernel is '1'
dilation = cv2.dilate(bw,kernel,iterations = 3) # total 3 iterations
plt.imshow(dilation,cmap='gray')
plt.show() #observe how the have been filled
```

### 1.2 Erosion
#### Erosion shrinks the connected sets of 1s of a binary image.
##### A pixel in the original image (either 1 or 0) will be considered 1 only if all the pixels under the kernel is 1, otherwise it is eroded (made to zero).

```python
kernel2 = np.ones((5,5),np.uint8)
erode = cv2.erode(img, kernel2,iterations = 3)
plt.imshow(erode, cmap='gray')
plt.show() #Observe how the gaps have enlarged 
```

### 1.3 Opening
##### Opening is the compound operation of erosion followed by dilation

```python
opening = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)
plt.imshow(opening, cmap='gray')
plt.show()
```

### 1.4 Closing
#### Closing is the compound operation of dilation followed by erosion

```python
closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel2)
plt.imshow(closing, cmap='gray')
plt.show()
```

### 1.5  Morphological Gradient
##### It is the difference between dilation and erosion of an image. The result image will look like the outline of the original image.

```python
gradient = cv2.morphologyEx(img, cv2.MORPH_GRADIENT, kernel2)
plt.imshow(gradient, cmap='gray')
plt.show()
```

### 1.6 Top Hat
#### It is the difference between input image and Opening of the image.

```python
tophat = cv2.morphologyEx(img, cv2.MORPH_TOPHAT, kernel)
plt.imshow(tophat, cmap='gray')
plt.show()
```

### 1.7 Black Hat
#### It is the difference between the closing of the input image and input image.

```python
blackhat = cv2.morphologyEx(img, cv2.MORPH_BLACKHAT, kernel)
plt.imshow(blackhat , cmap='gray')
plt.show()
```
